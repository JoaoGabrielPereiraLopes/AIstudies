```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
iris = datasets.load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
print(data.head())
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 3], iris.data[:, 2], c=iris.target)
ax.set(xlabel=iris.feature_names[3], ylabel=iris.feature_names[2])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
ax.set_facecolor('skyblue')
```

```python
X = np.array(iris.data[:, 3]).reshape(-1, 1)  # Largura da ptala
Y = np.array(iris.data[:, 2])  # Comprimento da ptala

# Criar e ajustar o modelo de regresso linear
modelo = LinearRegression()
modelo.fit(X, Y)

# Coeficientes da reta
m = modelo.coef_[0]  # Inclinao (coeficiente angular)
b = modelo.intercept_  # Intercepto

# Criar intervalo expandido para a reta (de X=0 at o mximo de X)
X_reg = np.linspace(0, max(X), 100).reshape(-1, 1)  # Expande at X=0
Y_reg = modelo.predict(X_reg)  # Calcula Y para os novos valores de X

# Criar o grfico
_, ax = plt.subplots()
scatter = ax.scatter(X, Y, c=iris.target)  # Plotar pontos reais
ax.set(xlabel=iris.feature_names[3], ylabel=iris.feature_names[2])

# Adicionar legenda para as classes
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)

# Criar a equao da reta com vrgulas no lugar de pontos
equation_text = f"y = {m:.3f}x + {b:.3f}".replace('.', ',')

# Plotar a reta expandida
ax.plot(X_reg, Y_reg, color="red", linestyle="--", label="Regresso Linear")

# Adicionar a equao da reta no ttulo do grfico
ax.set_title(f"Regresso Linear {equation_text}")

# Melhorar o layout
ax.legend(loc="upper left")
ax.grid(True)

# Exibir o grfico
plt.show()
```

```python
setosa_data = iris.data[target == 0]  # Setosa
versicolor_data = iris.data[target == 1]  # Versicolor
virginica_data = iris.data[target == 2]  # Virginica

# Criar o box plot horizontal
plt.figure(figsize=(8, 6))
box=plt.boxplot(
    [setosa_data[:, 2], versicolor_data[:, 2], virginica_data[:, 2]],  # Pegando apenas 'petal length (cm)'
    vert=True,  # Deixar horizontal
    patch_artist=True,  # Adicionar cor
    labels=["Setosa", "Versicolor", "Virginica"]  # Nome das espcies
)
colors = ['purple','cyan','gold']  # Definindo cores para cada caixa
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
# Configuraes do grfico
plt.title("Box Plot do Comprimento da Ptala por Espcie")
plt.ylabel("Comprimento da Ptala (cm)")
plt.grid(True)

# Exibir o grfico
plt.show()

```
